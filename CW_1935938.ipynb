```python
# CELL3 – 5 marks
################################################################################################################################
#  libraries needed for the coursework
%pylab inline
import pandas as pd
from pandas import Series, DataFrame
from scipy import stats
import statsmodels.api as sm
################################################################################################################################
ssheet=pd.read_excel('2018-QS-World-University-Rankings-Top200.xlsx')
ssheet.name= 'QS-UK-rankings'
ukonly=ssheet['Location'] == 'United Kingdom'
ukstats1=ssheet[ukonly].reset_index()
ukstats1.index=ukstats1.index+1
ukstats1=ukstats1.drop(['index','Location','Rank'],axis=1)

unroundedmean=ukstats1['Academic Reputation'].mean()
roundedmean=round(unroundedmean, 2)
print('Mean:', roundedmean)

unroundedmin=ukstats1['Academic Reputation'].min()
roundedmin=round(unroundedmin, 2)
print('Min:', roundedmin)

unroundedmax=ukstats1['Academic Reputation'].max()
roundedmax=round(unroundedmax, 2)
print('Max:', roundedmax)

unroundedstd=ukstats1['Academic Reputation'].std()
roundedstd=round(unroundedstd, 2)
print('Standard Deviation:', roundedstd)

zscore95p=1.960
mofe=zscore95p*(unroundedstd/sqrt(len(ukstats1)))
rmofe=round(mofe,2)
print('95% Confidence Interval: ', roundedmean, '±',rmofe, '. So the mean range at 95% Confidence Interval is ', round(roundedmean-rmofe, 2), 'to', roundedmean+rmofe)
```

```python
# CELL5 – 7 marks
################################################################################################################################
#  libraries needed for the coursework
%pylab inline
import pandas as pd
from pandas import Series, DataFrame
from scipy import stats
import statsmodels.api as sm
################################################################################################################################
ssheet=pd.read_excel('2018-QS-World-University-Rankings-Top200.xlsx')

usonly=ssheet['Location'] == 'United States'
usstats=ssheet[usonly]
uscpfmean=usstats['Citations per Faculty'].mean()
uscpfstd=usstats['Citations per Faculty'].std()
usciun=1.96*(uscpfstd/sqrt(len(usstats)))

nlonly=ssheet['Location'] == 'Netherlands'
nlstats=ssheet[nlonly].reset_index()
nlcpfmean=nlstats['Citations per Faculty'].mean()
nlcpfstd=nlstats['Citations per Faculty'].std()
nlciun=1.96*(nlcpfstd/sqrt(len(nlstats)))

mdif=round(uscpfmean-nlcpfmean,2)
cidi=round(usciun-nlciun,2)


plotdata=pd.DataFrame({'cpf':[uscpfmean, nlcpfmean, mdif]}, index=['United States', 'Netherlands', 'Difference'])
plotdata['cpf'].plot(kind="bar", title="test", yerr=[usciun,nlciun,cidi])

plt.title("Difference of mean and confidence interval level (CI) of Citations per Faculty between USA and Netherlands")
plt.xlabel("Country")
plt.ylabel("Mean Citations per Country and 95% CI")
print("There is a mean difference of ", abs(mdif), " citations. The confidence interval difference is ", abs(cidi), ", although the US has ", len(usstats), " universities while the Netherlands has ", len(nlstats))
print("Therefore while they have similar mean numbers, both reference so many sources, that it seems they have not done much primary research, but could be lower for Netherlands (More Scattered 95% Confidence Interval Level). I concluded that the Netherlands cite less per faculty than the US.")

```
